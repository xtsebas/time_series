```python
import pandas as pd
import matplotlib.pyplot as plt

# DataFrame con todas las m√©tricas
data = [
    # LSTM
    ["Gasolina Regular", "Importaci√≥n", "LSTM Modelo 1 (10 unidades, 1 lag)", 25.7735, 187981.4991, 232851.9702],
    ["Gasolina Regular", "Importaci√≥n", "LSTM Modelo 2 (32 unidades, 3 lags)", 26.1858, 190572.9850, 237250.6665],
    ["Gasolina Regular", "Consumo",      "LSTM Modelo 1 (10 unidades, 1 lag)",  8.6899,  65632.2055,  81852.6028],
    ["Gasolina Regular", "Consumo",      "LSTM Modelo 2 (32 unidades, 3 lags)", 10.4382,  64602.6660,  82196.5943],
    ["Diesel",           "Importaci√≥n", "LSTM Modelo 1 (10 unidades, 1 lag)", 22.5482, 256455.4020, 327089.1603],
    ["Diesel",           "Importaci√≥n", "LSTM Modelo 2 (32 unidades, 3 lags)", 22.7767, 258468.1658, 331317.7248],
    ["Diesel",           "Consumo",      "LSTM Modelo 1 (10 unidades, 1 lag)",  8.9096, 107676.8047, 138686.8852],
    ["Diesel",           "Consumo",      "LSTM Modelo 2 (32 unidades, 3 lags)",  7.9607,  94983.8151, 119506.7344],
    ["Gasolina Superior","Importaci√≥n", "LSTM Modelo 1 (10 unidades, 1 lag)",  7.9607,  94983.8151, 119506.7344],
    ["Gasolina Superior","Importaci√≥n", "LSTM Modelo 2 (32 unidades, 3 lags)",  7.9607,  94983.8151, 119506.7344],
    ["Gasolina Superior","Consumo",      "LSTM Modelo 1 (10 unidades, 1 lag)",  7.9607,  94983.8151, 119506.7344],
    ["Gasolina Superior","Consumo",      "LSTM Modelo 2 (32 unidades, 3 lags)",  7.9607,  94983.8151, 119506.7344],
    # Prophet & ARIMA
    ["Gasolina Regular", "Consumo",      "Prophet",      4.63,    37299,   45437],
    ["Gasolina Regular", "Importaci√≥n",  "Prophet",      8.61,    75015,  100269],
    ["Gasolina Superior","Consumo",      "Prophet",     11.46,    67130,   78780],
    ["Gasolina Superior","Importaci√≥n",  "ARIMA",       11.67,    70157,   96525],
    ["Diesel",           "Consumo",      "Prophet",      5.31,    66437,   85326],
    ["Diesel",           "Importaci√≥n",  "Prophet",     10.32,   133409,  161644],
]

columns = ["Combustible", "Escenario", "Modelo", "MAPE", "MAE", "RMSE"]
df = pd.DataFrame(data, columns=columns)

combustibles = ["Gasolina Regular", "Gasolina Superior", "Diesel"]
escenarios = ["Consumo", "Importaci√≥n"]

# M√©tricas a graficar por separado
for metric in ["MAPE", "MAE"]:
    for fuel in combustibles:
        for scenario in escenarios:
            subset = df[(df["Combustible"] == fuel) & (df["Escenario"] == scenario)]
            if subset.empty:
                continue

            plt.figure()
            plt.bar(subset["Modelo"], subset[metric])
            plt.ylabel(metric)
            plt.title(f"{fuel} - {scenario}: {metric}")
            plt.xticks(rotation=45, ha="right")
            plt.tight_layout()
            plt.show()
```

```python
for fuel in combustibles:
    for scenario in escenarios:
        subset = df[(df["Combustible"] == fuel) & (df["Escenario"] == scenario)]
        if subset.empty:
            continue

        plt.figure()
        plt.bar(subset["Modelo"], subset["RMSE"])
        plt.ylabel("RMSE")
        plt.title(f"{fuel} - {scenario}: RMSE")
        plt.xticks(rotation=45, ha="right")
        plt.tight_layout()
        plt.show()
```
